### **Modelo**

Definimos nuestro modelo

Estas capas conforman nuestra primera parte del modelo. La primera (Conv2D) realiza convoluciones para extraer características, max pooling reduce la dimensionalidad, y batch normalization estabiliza y acelera el entrenamiento.

**Capa Conv2D 1**:

- input_shape=(32, 32, 3): Esto especifica la forma de los datos de entrada a la red. En este caso, se espera que las imágenes de entrada tengan un tamaño de 32x32 píxeles con 3 canales de color (RGB).
- kernel_size=(3, 3): El tamaño de la ventana (filtro) que se desliza sobre la entrada. En este caso, es un filtro 3x3.
- filters=32: Indica el número de filtros (también conocidos como kernels o características) que se aplicarán. En este caso, hay 32 filtros.
- activation='relu': La función de activación Rectified Linear Unit (ReLU) se aplica después de la convolución. ReLU es común en redes neuronales convolucionales y ayuda a introducir no linealidades en la red.


**Capa MaxPooling2D**:

- pool_size=(2, 2): Realiza un submuestreo al seleccionar el valor máximo de una ventana 2x2. Esto reduce la dimensionalidad espacial y ayuda a reducir el costo computacional en las capas posteriores de la red.
- Nos extrae las características más importantes y reduce el tamaño de la representación.

**Capa BatchNormalization**:

- Normaliza la activación de cada filtro en una capa para tener media cercana a cero y desviación estándar cercana a uno. Esto ayuda a estabilizar y acelerar el entrenamiento, y mejora el rendimiento general de la red.

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
```

En el siguiente conjunto de capas, hemos decidido añadir una capa convolucional más pero aumentando el número de filtros. Esto ayuda a la red a aprender representaciones más complejas y abstractas.

Además, añadimos otra capa más MaxPooling2D y otra de BatchNormalization.

```python
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
```

Aplicamos el mismo pensamiento que hemos explicado anteriormente, dos capas convolucionales más aumentando el número de filtros.

Además, añadimos otra capa más MaxPooling2D y otra de BatchNormalization.

```python
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
```

Este conjunto de capas consecutivas (convolución, normalización y submuestreo) forma un patrón común en las CNNs, ayudando a la red a aprender y extraer características de las imágenes de entrada de manera progresiva y jerárquica.

En las capas iniciales, los filtros pueden capturar detalles simples como bordes, texturas, y colores básicos. A medida que avanzamos en capas más profundas, los filtros tienden a aprender representaciones más abstractas y de alto nivel.

Esto nos llevó a pensar que aumentar el número de filtros en las capas posteriores (32, 64, 128) permite que la red capture características más complejas y abstractas a medida que profundiza en las capas.

Las capas que añadimos a continuación reduccen la dimensionalidad para pasar de las representaciones espaciales aprendidas por las capas convolucionales a una representación plana que pueda ser utilizada por capas totalmente conectadas (densas). Analicemos cada capa:

**Capa Flatten**:

Esta capa transforma la salida de las capas convolucionales y de reducción de dimensionalidad en un vector unidimensional. Esto es necesario antes de pasar a capas totalmente conectadas (Dense) que requieren una entrada unidimensional.

**Capa Dense (256 neuronas)**:

- Añade una capa totalmente conectada con 256 neuronas (o unidades).
- La función de activación ReLU se aplica después de la operación lineal en cada neurona. ReLU introduce no linealidades y ayuda a que la red pueda aprender patrones más complejos.

**Capa Dropout (0.5)**:

Agrega una capa de Dropout con una tasa del 50%. Dropout es una técnica de regularización que ayuda a prevenir el sobreajuste al "apagar" aleatoriamente un porcentaje de neuronas durante el entrenamiento. Esto ayuda a que la red no dependa excesivamente de ciertas conexiones y mejora la generalización.

**Capa Dense (128 neuronas)**:

Añade otra capa totalmente conectada con 128 neuronas y función de activación ReLU.

**Capa Dropout (0.5)**:

Otra capa Dropout con una tasa del 50%.

**Capa Dense (128 neuronas)**:

Agrega otra capa totalmente conectada con 64 neuronas y función de activación ReLU.

**Capa Dense de Salida con 10 Neuronas (Número de Clases) y Activación Softmax**:

Última capa densa con 10 neuronas, que corresponde al número de clases en nuestro problema. Utiliza la función de activación softmax (problemas de clasificación multiclase). Produce probabilidades para cada clase y es útil en la salida de la red para la clasificación.

```python
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

###**Compilación del modelo**

Ahora compilaremos nuestro modelo

Evaluemos los parámetros:

**Loss Function (Función de Pérdida)**
- loss='categorical_crossentropy': Esta es la función de pérdida que utilizaremos durante el entrenamiento del modelo. En este caso, utilizamos *categorical_crossentropy*, que es comúnmente utilizada en problemas de clasificación con múltiples clases. Es adecuada cuando se trabaja con salidas one-hot-encoded, como en nuestro caso, cuando las etiquetas de destino son vectores categóricos binarios.

**Optimizer (Optimizador)**
- optimizer=adam(): Este es el optimizador que utilizaremos para minimizar la función de pérdida durante el entrenamiento. En este caso, se utiliza el optimizador Adam. Adam (Adaptive Moment Estimation) adapta dinámicamente la tasa de aprendizaje para cada parámetro de la red. A diferencia de los métodos con tasas de aprendizaje fijas, Adam ajusta la tasa de aprendizaje para cada parámetro individualmente en función de sus históricos gradientes. Esto permite un aprendizaje más rápido en dimensiones donde los gradientes son más grandes y más lento en dimensiones con gradientes más pequeños.

Probamos también a poner el optimizador SGD, pero el que mejor ha hecho que aprenda nuestra red ha sido Adam.

**Metrics (Métricas de Evaluación)**

- metrics=['accuracy']: Durante el entrenamiento y la evaluación del modelo, se hará un seguimiento de la precisión (accuracy) como métrica adicional. La precisión es una medida común en problemas de clasificación y representa la fracción de instancias correctamente clasificadas sobre el total.

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

###**Entrenamiento del modelo**

Entrenamos nuestro modelo

El 573 significa que estamos en el lote (*batch*) número 573 de un total de 573 lotes en la época actual. Este número viene determinado por el *batch_size* que hemos elegido (128). En la mayoría de los casos, los lotes se eligen aleatoriamente durante el entrenamiento de modelos de aprendizaje profundo. Este enfoque se conoce como "entrenamiento por lotes aleatorios" o "aleatorización de lotes". La idea es que cada época del entrenamiento se compone de múltiples lotes, y en cada época, los lotes se eligen aleatoriamente a partir del conjunto de datos de entrenamiento.

```python
history = model.fit(train_images, train_labels_one_hot, epochs=15, batch_size=128, validation_data=(test_images, test_labels_one_hot))
```

A continuación, realizamos la predicción de la que hablábamos.

Como hemos explicado anteriormente, según el resultado de nuestra precisión con los datos de test, nuestra red acertará un 92% de las veces. Teniendo 8 fotos, por lo tanto, nos debería acertar por lo menos 7 fotografías (8*0.92 = 7.36 -> aprox 7 fotos).

```python
# Seleccionar siete índices de imágenes de prueba (ajusta según sea necesario)
sample_indices = [11, 14, 20, 35, 42, 55, 68, 79]
class_names = ['clase_{}'.format(i) for i in range(100)]

# Configurar la visualización de las imágenes en una cuadrícula de 2x4
plt.figure(figsize=(12, 8))

for i, sample_index in enumerate(sample_indices, 1):
    # Seleccionar la imagen y etiqueta correspondiente
    sample_image = test_images[sample_index]
    sample_label = test_labels_one_hot[sample_index]

    # Obtener la posición de la clase real (usando np.where)
    real_class_index = np.where(sample_label == 1)[0][0]

    # Expandir las dimensiones para que coincidan con las expectativas del modelo
    sample_image = np.expand_dims(sample_image, axis=0)

    # Obtener la predicción del modelo
    prediction = model.predict(sample_image)
    predicted_label = np.argmax(prediction)

    # Mostrar la imagen
    plt.subplot(2, 4, i)
    plt.imshow(test_images[sample_index])
    plt.title(f'Real: {class_names[real_class_index]}\nPredicha: {class_names[predicted_label]}')
    plt.axis('off')

plt.show()
```

Los resultados que obtenemos son los que esperábamos. Ha acertado 7 de 8 imágenes.

El sobreentrenamiento se puede deber a varias razones:

**Hiperparámetros no óptimos**:

Los hiperparámetros del modelo, como la tasa de aprendizaje, la tasa de abandono (dropout), o la complejidad del modelo, pueden no estar ajustados correctamente. Estos hiperparámetros pueden afectar la capacidad del modelo para generalizar.

**Cantidad limitada de datos**:

Si el conjunto de datos de entrenamiento es pequeño, el modelo puede aprender rápidamente a ajustarse a esos datos, pero no generalizará bien a datos no vistos.

**Falta de regularización**:

Puede ser útil agregar técnicas de regularización, como la regularización L2, para ayudar a prevenir el sobreajuste.
